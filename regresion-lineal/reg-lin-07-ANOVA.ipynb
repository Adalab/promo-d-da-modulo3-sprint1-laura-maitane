```python
# Tratamiento de los datos
# ========================
import numpy as np
import pandas as pd

# Librer铆as para la visualizaci贸n de los datos
# ============================================ 
import matplotlib.pyplot as plt 
import seaborn as sns

# Estad铆sticos
# ============
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

# Establecer preferencias de visualizaci贸n
# ========================================
plt.rcParams["figure.figsize"] = (10,8) 
```

```python
df = pd.read_csv("datos/soci_econ_country_profiles_estandarizado.csv", index_col = 0)
df.sample(5)
```

```python
# iniciamos la ANOVA para la variable respuesta normalizada 'restaurant_price_index_normal'
lm = ols('''restaurant_price_index_normal ~ region + population_2017 + employment_agriculture + employment_industry + employment_services + education +
          quality_of_life_index + purchasing_power_index + safety_index + cost_of_living + consumer_price_index + human_capital_index +
          inflation + urban_population + population_age_distribution_0_14 + population_age_distribution_60_plus + international_migrant_stock''', 
          data=df).fit()

# resultados obtenidos en el ANOVA
sm.stats.anova_lm(lm)
```

```python
# sacamos el resumen de la recta que ajusta nuestro modelo
lm.summary()
```
