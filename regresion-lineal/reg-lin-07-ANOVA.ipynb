```python
# Tratamiento de los datos
# ========================
import numpy as np
import pandas as pd

# Librerías para la visualización de los datos
# ============================================
import matplotlib.pyplot as plt 
import seaborn as sns

# Estadísticos
# ============
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

plt.rcParams["figure.figsize"] = (10,8) 
```

```python
df = pd.read_csv("datos/soci_econ_country_profiles_estandarizado.csv", index_col = 0)
df.sample(5)
```

```python
# iniciamos la ANOVA para la variable respuesta normalizada 'restaurant_price_index_normal'
lm = ols('''restaurant_price_index_normal ~ region + population_2017 + employment_agriculture +
       employment_industry + employment_services + education +
       quality_of_life_index + purchasing_power_index + safety_index +
       cost_of_living + consumer_price_index + human_capital_index +
       inflation + urban_population + population_age_distribution_0_14 +
       population_age_distribution_60_plus + international_migrant_stock''', 
         data=df).fit()
# resultados obtenidos en el ANOVA
sm.stats.anova_lm(lm)
```

```python
# sacamos el resumen de la recta que ajusta nuestro modelo
lm.summary()
```

CONCLUSIONES:

- El R-Square de nuestros datos es muy alto, llegando a 0.96, es decir, los datos sugieren que nuestro modelo de predición es muy fiable porque explica el 96% de la variación de la variable respuesta. Sin embargo, estas conclusiones son fictícias ya que nuestros datos no siguen una distribución normal. 
  
- Las variables que afectan la variable respuesta (pvalue <= 0.05) en nuetros datos son: 'region', 'purchasing_power_index', 'cost_of_living', 'human_capital_index', 'inflation' y 'urban_population'. Estas son las que deberemos usar para el modelo de regresión lineal. 
  
- De las columnas que afectan nuestra VR sacamos los coef: 'region' (con distintos valores), 'purchasing_power_index' (coef: -0.2593), 'cost_of_living' (coef 0.6338), 'human_capital_index' (coef 0.2686), 'inflation' (coef -0.0758) y 'urban_population' (coef 0.2021). En el caso de la variable 'purchasing_power_index' (coef: -0.2593), por cada unidad que aumenta 'purchasing_power_index', la variable respuesta disminuye 0.2593.
